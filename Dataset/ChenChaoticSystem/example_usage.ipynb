```python
import os
import pandas as pd
from save_dataset import save_dataset
import matplotlib.pyplot as plt
```

```python
# set the parameters

P = 4  # number of coupled oscillators per folder (entities)
num_folders = 5  # number of the run
num_files_per_folder = P  # one CSV per entity

t_max = 20.0       # simulation time
dt = 0.01
t_switch = 10.0    # switch time

a, b = 35.0, 2.8
c1, c2 = 24.0, 27.0
lambda_env = 1.0
epsilon1_on, epsilon2_on = 0.5, 21
epsilon1_off, epsilon2_off = 0.0, 0.0

# Output path for test
output_path = r"path_To_Output\Normal"
```

```python
save_dataset(output_path, num_folders, num_files_per_folder,
             P, t_max, dt, t_switch,
             a, b, c1, c2, lambda_env,
             epsilon1_on, epsilon2_on,
             epsilon1_off, epsilon2_off,
             mode="normal")
```

```python
# List all folders
folders = os.listdir(output_path)
print("The number of folders created:", len(folders))

# Count total CSV files
total_csv_files = sum(len(os.listdir(os.path.join(output_path, f))) for f in folders)
print("The total number of CSV files created:", total_csv_files)

# Load first CSV of the first folder as a preview
first_folder = os.path.join(output_path, folders[0])
csv_files = os.listdir(first_folder)
df = pd.read_csv(os.path.join(first_folder, csv_files[0]))

print("Preview of first CSV:")
print(df.head())
```

```python
# Path to first folder and CSV
first_folder = os.path.join(output_path, folders[0])
csv_files = os.listdir(first_folder)
csv_path = os.path.join(first_folder, csv_files[0])

# Load data
df = pd.read_csv(csv_path)

# Plot x
plt.figure(figsize=(10, 3))
plt.plot(df['time'], df['x'], color='blue', label='x')
plt.axvline(x=t_switch, color='red', linestyle='--', label='t_switch')
plt.xlabel('Time')
plt.ylabel('x')
plt.title('x vs Time with t_switch')
plt.legend()
plt.grid(True)
plt.show()

# Plot y
plt.figure(figsize=(10, 3))
plt.plot(df['time'], df['y'], color='orange', label='y')
plt.axvline(x=t_switch, color='red', linestyle='--', label='t_switch')
plt.xlabel('Time')
plt.ylabel('y')
plt.title('y vs Time with t_switch')
plt.legend()
plt.grid(True)
plt.show()

# Plot z
plt.figure(figsize=(10, 3))
plt.plot(df['time'], df['z'], color='green', label='z')
plt.axvline(x=t_switch, color='red', linestyle='--', label='t_switch')
plt.xlabel('Time')
plt.ylabel('z')
plt.title('z vs Time with t_switch')
plt.legend()
plt.grid(True)
plt.show()
```
